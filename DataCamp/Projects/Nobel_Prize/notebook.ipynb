The Nobel Prize has been among the most prestigious international awards since 1901. Each year, awards are bestowed in chemistry, literature, physics, physiology or medicine, economics, and peace. In addition to the honor, prestige, and substantial prize money, the recipient also gets a gold medal with an image of Alfred Nobel (1833 - 1896), who established the prize.

![](Nobel_Prize.png)

The Nobel Foundation has made a dataset available of all prize winners from the outset of the awards from 1901 to 2023. The dataset used in this project is from the Nobel Prize API and is available in the `nobel.csv` file in the `data` folder.

In this project, you'll get a chance to explore and answer several questions related to this prizewinning data. And we encourage you then to explore further questions that you're interested in!

```python
# Loading in required libraries
import pandas as pd
import seaborn as sns
import numpy as np
    
# Start coding here!
```

```python
df=pd.read_csv('data/nobel.csv')
df.head()
```

```python
df.astype({'full_name': 'string','sex':'string'}).dtypes
```

```python
vcs=df.value_counts(df['sex'])
tg=vcs[vcs==vcs.max()]
top_gender=tg.index[0]
```

```python
top_gender
```

```python
vcbc=df.value_counts(df['birth_country'])
tc=vcbc[vcbc==vcbc.max()]
top_country=tc.index[0]
```

```python
top_country
```

```python
df_usa=df.copy()[df['birth_country']=='United States of America']
df_usa['decade'] = (df_usa['year'] // 10) * 10
vcd=df.value_counts(df_usa['decade'])
mdu=vcd[vcd==vcd.max()]
max_decade_usa=int(mdu.index[0])
```

```python
max_decade_usa
```

```python
df_fem=df.copy()
df_fem['decade']=(df_fem['year'] // 10) * 10
df_fem['fem']=(df_fem['sex']=='Female')
df_fem=df_fem.groupby(['decade','category'],as_index=False)['fem'].mean()
mfc=df_fem[df_fem['fem']==df_fem['fem'].max()]
max_female_dict={mfc.values[0][0]:mfc.values[0][1]}
```

```python
repeat_list=[]
# Individuals
df_ind=df.copy()[df['laureate_type']=='Individual']
ind=df_ind['full_name'].value_counts()
list_ind=ind[ind[:]>1]
for i in list_ind.index:
    repeat_list.append(i)
# Organizations
df_org=df.copy()[df['laureate_type']=='Organization']
org=df_org['full_name'].value_counts()
list_org=org[org[:]>1]
for i in list_org.index:
    repeat_list.append(i)
```
